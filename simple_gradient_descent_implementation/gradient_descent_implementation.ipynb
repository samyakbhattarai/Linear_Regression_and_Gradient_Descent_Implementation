### Y = MX + C + noise = Y = 2x + 1 ; m = 2 and c = 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Assigning values to X and Y and creating noise

```python
np.random.seed(42)

X = np.linspace(0, 10, 100)

true_m, true_c = 2, 1

Y = true_m*X + true_c + np.random.normal(0, 2, X.shape[0])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,Y, color="blue", alpha=0.6, label="Data Points")
plt.legend()
plt.show()
```

# Loss Function

```python
loss_history=[]
# L(w,c) = 1/n * sum((ypred-y)^2)
# Pd(L(w)) = 2/n * X * (ypred - y)
# Pd(L(c)) = 2/n * (ypred - y)

m,c = 0,0
iterations = 2000
learning_rate = 0.01
```

```python
for i in range(iterations):
    Ypred = m * X + c
    grad_m = 2/len(X) * np.sum(X * (Ypred - Y))
    grad_c = 2/len(X) * np.sum(Ypred - Y)

    m = m - learning_rate * grad_m
    c = c - learning_rate * grad_c

    loss = np.mean((Ypred - Y) ** 2)
    loss_history.append(loss)

    if(i % 100 == 0):
        print(f"Loss: {loss} for M: {m} and C: {c}")
    
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,Y, color="blue", alpha=0.6, label="Data Points")
plt.plot(X, m*X + c, color="red", label=f"Model Fit {m:.2f}x + {c:.2f}")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(iterations),loss_history)
plt.xlabel("Iterations")
plt.ylabel("Loss History")
plt.ylim(0, 10)
plt.xlim(0,10)
plt.show()
```
