```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
model_dataset = df[['price','area']]
```

```python
to_train = model_dataset
```

```python
m = 100
X = to_train['area']/1000
```

```python
Y = to_train['price']/1000000
```

```python
plt.figure(figsize=(8,6))
plt.scatter(sorted(X),sorted(Y),color="red",alpha=0.6, label="Training Data")
plt.xlabel("House Size(1000 sq ft)")
plt.ylabel("Price of House($1,000,000)")
plt.title("House Prediction Model")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
